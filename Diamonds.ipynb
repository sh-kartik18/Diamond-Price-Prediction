```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
categorical_cols = ['cut', 'color', 'clarity']
numerical_cols = ['carat', 'depth', 'table', 'x', 'y', 'z']
```

```python
preprocessor = ColumnTransformer([
    ('num', StandardScaler(), numerical_cols),
    ('cat', OneHotEncoder(drop='first'), categorical_cols)
])
```

```python
X = df.drop('price', axis=1)
y = df['price']
```

```python
X_preprocessed = preprocessor.fit_transform(X)
```

```python
X_preprocessed
```

```python
print(df['cut'].unique())    
print(df['color'].unique())   
print(df['clarity'].unique()) 
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.3, random_state=0)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor

models = {
    "Linear Regression": LinearRegression(),
    "Ridge Regression": Ridge(),
    "Lasso Regression": Lasso(),
    "Decision Tree Regression": DecisionTreeRegressor(),
    "Random Forest Regression": RandomForestRegressor(),
    "Support Vector Regression": SVR(kernel='linear', C=1.0, epsilon=0.2, gamma='scale'),
    "Neural Network Regression": MLPRegressor(hidden_layer_sizes=(128, 64), activation='relu', solver='adam',max_iter=1000, random_state=42)
}
```

```python
#Master Copy
```

```python
results = {}
for model_name, model in models.items():
    model.fit(X_train, y_train) 
    y_pred = model.predict(X_test)  
    
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    
    results[model_name] = {"MAE": mae, "MSE": mse, "R^2": r2}
    
results_df = pd.DataFrame(results).T
results_df = results_df.sort_values(by="R^2", ascending=False)

if 'Model' not in results_df.columns:
    results_df.reset_index(inplace=True)
    results_df.rename(columns={'index': 'Model'}, inplace=True)
results_df.set_index('Model', inplace=True) 

print(results_df)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))

results_df['R^2'].sort_values(ascending=False).plot(kind='bar', color='coral', ax=axes[0])
axes[0].set_title("R² Score Comparison")
axes[0].set_xlabel("Model")
axes[0].set_ylabel("R²")

results_df['MAE'].sort_values(ascending=False).plot(kind='bar', color='salmon', ax=axes[1])
axes[1].set_title("Mean Absolute Error (MAE) Comparison")
axes[1].set_xlabel("Model")
axes[1].set_ylabel("MAE")

results_df['MSE'].sort_values(ascending=False).plot(kind='bar', color='lightgreen', ax=axes[2])
axes[2].set_title("Mean Squared Error (MSE) Comparison")
axes[2].set_xlabel("Model")
axes[2].set_ylabel("MSE")

fig.suptitle(f"Model Comparison for df(Master Copy)", fontsize=16)

plt.tight_layout()

plt.show()
```

```python
results_df['Source'] = 'Master Copy'
results_df_min['Source'] = 'Min'
results_df_max['Source'] = 'Max'
results_df_mean['Source'] = 'Mean'
results_df_sd['Source'] = 'Standard Deviation'

compiled_results = pd.concat(
    [results_df, results_df_min, results_df_max, results_df_mean, results_df_sd],axis=0).reset_index()

best_mae = compiled_results.loc[compiled_results['MAE'].idxmin()]  
best_mse = compiled_results.loc[compiled_results['MSE'].idxmin()]  
best_r2 = compiled_results.loc[compiled_results['R^2'].idxmax()]   

summary = pd.DataFrame({
    'Metric': ['MAE', 'MSE', 'R²'],
    'Best Model': [best_mae['Model'], best_mse['Model'], best_r2['Model']],
    'Best DataFrame': [best_mae['Source'], best_mse['Source'], best_r2['Source']],
    'Value': [best_mae['MAE'], best_mse['MSE'], best_r2['R^2']]
})

print(summary)
```
